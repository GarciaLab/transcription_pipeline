```python
def simulate_mcp_binding(
    loops_time,          # numpy array of # of MS2 loops at each timepoint
    deltat=0.1,          # time step
    r_in=50.0,           # MCP entry rate (Poisson mean per unit time)
    r_out= 1.0,
    k_on=10,
):
    n_steps = len(loops_time)

    # Initialize arrays to store outputs
    mcp_in_volume = np.zeros(n_steps, dtype=int)
    occupied_loops = np.zeros(n_steps, dtype=int)
    unbound_loops = np.zeros(n_steps, dtype=int)

    # Keep track of the number of free MCP molecules in the binding volume
    free_mcp = r_in/r_out  # initial number of free MCP molecules
    bound_loops = 0

    for t in range(n_steps):
        total_loops = loops_time[t]

        # Adjust bound/unbound if loop count changes
        if total_loops > bound_loops + (unbound_loops[t-1] if t > 0 else 0):
            # New loops appear → they start unbound
            new_loops = total_loops - (bound_loops + (unbound_loops[t-1] if t > 0 else 0))
            currently_unbound = (unbound_loops[t-1] if t > 0 else 0) + new_loops
        else:
            # Loops disappear → remove from unbound first, then bound
            lost_loops = (bound_loops + (unbound_loops[t-1] if t > 0 else 0)) - total_loops
            prev_unbound = unbound_loops[t-1] if t > 0 else 0
            if lost_loops <= prev_unbound:
                currently_unbound = prev_unbound - lost_loops
            else:
                lost_from_bound = lost_loops - prev_unbound
                bound_loops = max(0, bound_loops - lost_from_bound)
                currently_unbound = 0

        # MCP entering the volume (Poisson process)
        m_in = np.random.poisson(r_in * deltat)
        free_mcp += m_in

        # MCP binding to loops (2 MCP per loop)
        if currently_unbound > 0 and free_mcp >= 2:
            # Probability of binding
            p_bind = min(k_on * free_mcp * deltat, 1.0)
            m_out_loops = np.random.binomial(currently_unbound, p_bind)
            m_out = min(free_mcp // 2, m_out_loops) * 2  # cap by available MCP
            loops_bound_now = m_out // 2

            # Update counts
            free_mcp -= m_out
            bound_loops += loops_bound_now
            currently_unbound -= loops_bound_now
        # Outflow of MCP (Poisson process)
        m_out_flux = np.random.poisson(r_out * free_mcp * deltat)
        free_mcp -= m_out_flux
        # Store results
        mcp_in_volume[t] = free_mcp
        occupied_loops[t] = bound_loops
        unbound_loops[t] = currently_unbound

        # Sanity check (should always hold)
        # assert occupied_loops[t] + unbound_loops[t] == total_loops, \
        #     f"Loop count mismatch at t={t}"

    return mcp_in_volume, occupied_loops, unbound_loops


# Example usage
# np.random.seed(42)
time_points = 100
loops_time = np.random.randint(5, 15, size=time_points)   # Random # loops at each time
mcp_conc_time = np.linspace(0, 50, time_points)           # Just for example
deltat = 0.1

mcp_in_volume, occupied_loops, unbound_loops = simulate_mcp_binding(
    loops_time=loop_time_test, deltat=deltat, r_in=100.0, r_out = 1.0, k_on=1,
)

# Example plot
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.plot(occupied_loops, label="Occupied loops")
plt.plot(unbound_loops, label="Unbound loops")
plt.plot(loop_time_test, label="Total loops", linestyle='--', alpha=0.5, lw=3)
plt.plot(mcp_in_volume, label="Free MCP molecules")
plt.xlabel("Time step")
plt.ylabel("Number")
plt.legend()
plt.show()
```
